```python
from scipy import stats

# Calculate the mean, median, and mode for the 'Price' column
mean_price = df['Price'].mean()
median_price = df['Price'].median()
mode_price = df['Price'].mode()[0]  # mode() returns a Series, take the first mode if multiple
Std = df['Price'].std()
range_value = df['Price'].max() - df['Price'].min()
max_value = df['Price'].max()
min_value = df['Price'].min()
q1 = df['Price'].quantile(0.25)
q3 = df['Price'].quantile(0.75)
iqr = q3 - q1

# Print the results
print(f"Mean Price: {mean_price}")
print(f"Median Price: {median_price}")
print(f"Mode Price: {mode_price}")
print(f"Standard Deviation: {Std}")
print(f"Max Value: {max_value}")
print(f"Min Value: {min_value}")
print(f"Range: {range_value}")
print(f"Q1: {q1}")
print(f"Q3: {q3}")
print(f"IQR: {iqr}")
```

```python
average_price_by_location = df.groupby('SquareFootage')['Price'].max().reset_index()

# Set up the figure size
plt.figure(figsize=(10, 6))

# Bar plot of average price by location
sns.barplot(data=average_price_by_location, x='SquareFootage', y='Price', palette='viridis')
plt.title('Bar Plot of Average Price by Location')
plt.xlabel('Location')
plt.ylabel('Average Price')

# Show the plot
plt.show()
```

```python
average_price_by_square_footage_bin = df.groupby('SquareFootage')['Price'].mean().reset_index()

# Set up the figure size
plt.figure(figsize=(10, 6))

# Bar plot of average price by square footage bin
sns.barplot(data=average_price_by_square_footage_bin, x='SquareFootage', y='Price', palette='viridis')
plt.title('Bar Plot of Average Price by Square Footage')
plt.xlabel('Square Footage ')
plt.ylabel('Average Price')
```

```python
average_price_by_Bedrooms = df.groupby('Bedrooms')['Price'].mean().reset_index()

# Set up the figure size
plt.figure(figsize=(10, 6))

# Bar plot of average price by square footage bin
sns.barplot(data=average_price_by_Bedrooms, x='Bedrooms', y='Price', palette='viridis')
plt.title('Bar Plot of Average Price by Number of Bedrooms')
plt.xlabel('Bedrooms ')
plt.ylabel('Average Price')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Price'])
plt.title('Box Plot of House Prices')
plt.xlabel('Price')
plt.show()
```

```python
price_data = df['Price'].dropna()

q1_val = price_data.quantile(0.25)
median_val = price_data.median()
q3_val = price_data.quantile(0.75)


# Create a box plot for the 'Price' column
plt.figure(figsize=(10, 6))
sns.boxplot(x=price_data,color='yellow')
plt.title('Box Plot of House Prices')
plt.xlabel('Price')

# Annotate the box plot with statistical values

plt.text(q1_val, 0.02, f'Q1: {q1_val}', ha='center', va='center', color='green', fontsize=15)
plt.text(median_val, 0.02, f'Median: {median_val}', ha='center', va='center', color='red', fontsize=15)
plt.text(q3_val, 0.02, f'Q3: {q3_val}', ha='center', va='center', color='green', fontsize=15)


plt.show()
```
